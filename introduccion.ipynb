FOR


```python
#Los for son parecidos a los if, pero tienen la sintaxis for variable in lista:. En este caso, 
# variable es la variable que va a ir cambiando, y lista es una lista de python (o un iterable que es parecido)
print("-Impresion de los elementos de una lista")

mi_lista=["img","python","java"]
for i in mi_lista:
    print(i)

#CALCULA LA SUMA DE LOS NUMEROS E IMPRIMIRLA
suma=0
mis_numeros=[5,8,22,43]
for i in mis_numeros:
    suma=suma+i

print("-La suma de los numeros es")
print(suma)
```

```python
#Cuando no tenemos una lista y queremos hacer un for "común" y que la variable que 
# cambia sea un número que va incrementándose, podemos utilizar la función range.
print("Un for de 0 a 3")
for i in range(4):
    print(i)

#EN PYTHON LOS INDICES COMIENZAN EN 0 Y POR ESO LOS RANGOS TAMBIEN

#TAMBIEN SE PUEDE COMENZAR EL RANGO EN OTRO VALOR EN LUGAR DE 0
print("- Un for de 2 a 5")
for j in range(2,6):
    print(j)
```

```python
#EJERCICIO: ESCRIBIR UN FOR PARA BUSCAR EL MAXIMO DE LA LISTA E IMPRIMIRLO
mis_numeros=[5,8,22,43,3,67,34,89,2]
maximo=0
#IMPLEMENTAR
for i in mis_numeros:
    if i>maximo:
        maximo=i

print("El maximo es:")
print(maximo)
```

```python
#EJERCICIO: ESCRIBIR UN FOR PARA BUSCAR EL MAXIMO D ELA LISTA E IMPRIMIR SU POSICION
mis_numeros=[5,8,22,43,3,67,34,89,2]
maximo=0
#IMPLEMENTAR
posicion=-1
for i in range(len(mis_numeros)):
    if mis_numeros[i]>maximo:
        maximo=mis_numeros[i]
        posicion=i
    
print("El maximo es:", maximo)
print("Su posicion es:", posicion)
```

FUNCIONES

```python
#ESTA FUNCION RECIBE DOS NUMEROS Y DEVULEVE SU SUMA
def sumar(a,b):
    return a+b

resultado=sumar(5,10)
print("La suma es:", resultado)

#ESTA FUNCION RECIBE UNA LISTA Y DEVUELVE LA SUMA DE SUS ELEMENTOS
def sumar_lista(lista_numeros):
    suma=0
    for i in lista_numeros:
        suma+=i
    return suma

resultado=sumar_lista([10,20,30,40])
print("La suma de la lista es:", resultado)
```

```python
#EJERCICIO 
#ESCRIBIR UNA FUNCION QUE RECIBA UNA LISTA Y UN VALOR
#Y DEUVEKVA LA CANTIDAD DE VECES QUE APARECE ESE VALOR EN LA LISTA

def contar_ocurrencias(lista, valor):
    contador=0
    #IMPLEMENTAR
    for i in lista:
        if i==valor:
            contador+=1
    return contador

lista=[1,4,2,3,5,1,4,2,3,6,1,7,1,3,5,1,1,5,3,2]
valor=1

ocurrencias=contar_ocurrencias(lista, valor)
print("El valor", valor, "aparece", ocurrencias, "veces en la lista")
```
