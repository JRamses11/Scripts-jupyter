INTEGRACION DE BIBLIOTECAS DE PYTHON

```python
import pandas as pd
from sklearn.datasets import load_diabetes
```

```python
#CARGAR EL CONJUNTO DE DATOS DE DIABETES
diabetes=load_diabetes()
data=pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
data['target']=diabetes.target

#MOSTRAR LAS PRIMERAS FILAS DEL DATAFRAME
print(data.head())

print(data.describe())
```

```python
import matplotlib.pyplot as plt
#HISTOGRAMA DE LA EDAD DE LOS PACIENTES
plt.figure(figsize=(10,6))
plt.hist(data['age'], bins=20, color='blue', edgecolor='black')
plt.title('Distribución de la Edad de los Pacientes')
plt.xlabel('Edad (normalizada)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
import seaborn as sns
#CREAR UN MAPA DE CALOR DE LAS CORRELACIONES ENTRE CARACTERÍSTICAS
correlation_matrix= data.drop('target', axis=1).corr()
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de Calor de las Correlaciones entre Características')
plt.show()
```

```python
#CREAR UN PAR DE GRAFICOS PARA VISUALIZAR RELACIONES ENTRE CARACTERISTICAS
sns.pairplot(data)
plt.show()
```
