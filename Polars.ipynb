# INTRODUCCION A POLARS

```python
#LIBRERIAS
import polars as pl # Para manejo de dataframes
from datetime import datetime
```

```python
df_titanic = pl.read_csv('titanic.csv') #FUNCION PARECIDA A read_csv DE PANDAS
df_titanic.head()
```

```python
#El método .tail() te permite ver las últimas $N$ filas de tu DataFrame (o LazyFrame). Es útil para 
# verificar si los datos se han cargado correctamente o si hay valores atípicos al final del conjunto.
df_titanic.tail()

#Te permite seleccionar una muestra aleatoria de filas de tu DataFrame (o LazyFrame).
df_titanic.sample(5)
```

```python
n_chars = 50

# df.columns
# ==============================================================================
print("="*n_chars)
print("df.columns: listado de columnas de un DataFrame")
print(df_titanic.columns)
print("")

# df.types
# ==============================================================================
print("="*n_chars)
print("df.types: tipos de datos de las columnas de un DataFrame")
print(df_titanic.dtypes)
print(f"Tipos de datos únicos: {set(df_titanic.dtypes)}")
print("")

# df.schema
# ==============================================================================
print("="*n_chars)
print("df.schema: Diccionario con las columnas y sus tipos de datos")
print(df_titanic.schema)
print("")

# df.shape
# ==============================================================================
print("="*n_chars)
print("df.shape : filas x columnas de un DataFrame")
print(df_titanic.shape)
print("")

# df.height
# ==============================================================================
print("="*n_chars)
print("df.height : número de filas de un DataFrame")
print(df_titanic.height)
print("")

# df.width
# ==============================================================================
print("="*n_chars)
print("df.width : número de columnas de un DataFrame")
print(df_titanic.width)
print("")
```

```python
#df.describe()
print("df.describe(): estadísticos descriptivos de las columnas numéricas")
display(df_titanic.describe())
```

# Valores nulos

```python
#Devuelve un DataFrame con el total de valores nulos por columna.
#df.null_count()
print("df.null_count(): conteo de valores nulos por columna")
display(df_titanic.null_count())
```

# Polars expressions

```python
# .select(): Permite seleccionar columnas específicas de un DataFrame o LazyFrame.
df_titanic.select(pl.col("Age"), pl.col("Fare"))
```

```python
#Por defecto, este método nos retorna un otro DataFrame (polars.internals.dataframe.frame.DataFrame)
type(df_titanic.select(pl.col("Age"), pl.col("Fare")))  # Devuelve un DataFrame
```

```python
# Selección de una única columna
# ==============================================================================
display(df_titanic.select("Age").head())

# Selección de múltiples columnas
# ==============================================================================
display(df_titanic.select(["Age","Fare"]).head())
```
