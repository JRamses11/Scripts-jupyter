```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
#CONFIGURACION DE ESTILO DE LAS VISUALIZACIONES
iris=load_iris()
data=pd.DataFrame(data=iris.data, columns=iris.feature_names)
data['species']=iris.target

sns.set(style="whitegrid", palette="muted", color_codes=True, font_scale=1.2, rc={"figure.figsize": (10, 6)})

#CARGAR EL DATASET DE IRIS
```

#ANALISIS DE DATOS

En este notebook, analizaremos un conjunto de datos del famoso dataset Iris (incluido dentro del paquete Seaborn)

```python
#Informacion general del dataset
data.info()
#Primeras filas del dataset
data.head()
#Visualizacion de la distribucion de las caracteristicas
plt.figure(figsize=(10, 6))
for i, column in enumerate(data.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribucion de {column}')
plt.tight_layout()
plt.show()
```
